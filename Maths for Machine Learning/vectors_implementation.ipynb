### Implementing vector operations in python:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sns.set()
```

Addition of two vectors

```python
vector_1 = np.asarray([0,0,2,3])
vecotr_2 = np.asarray([0,0,3,-2])
sum = vector_1 + vecotr_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units = "xy", angles = "xy", scale  =1, color = "b")
plt.quiver(0,0,3,-2, scale_units = "xy", angles = "xy", scale = 1, color = "b")
plt.quiver(0,0,5,1, scale_units = "xy", angles = "xy", scale = 1, color = "r")
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of two vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])

difference = vector_1 - vector_2
print(difference)
```

```python
plt.quiver(0,0,2,3, scale_units="xy", angles = "xy", scale = 1, color = "b")
plt.quiver(0,0,3,-2, scale_units = "xy", angles = "xy", scale = 1, color = "b")
plt.quiver(0,0,-1,5, scale_units = "xy", angles = "xy", scale = 1, color = "r")
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by a scalar

```python
vector_1 = np.asarray([0,0,2,3])
scalar = 2
product = scalar * vector_1
print(product)
```

```python
plt.quiver(0,0,2,3, scale_units = "xy", angles = "xy", scale = 1,color = "b")
plt.quiver(0,0,4,6, scale_units = "xy", angles = "xy", scale = 1,color = "r")
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Dot Product

```python
a = np.array([2,3])

b = np.array([4,4])

dot_product = np.dot(a,b)

print(dot_product)
```

Cross product

```python
a = np.array([2,3])

b = np.array([4,4])

dot_product = np.cross(a,b)

print(dot_product)
```

```python
c = np.array([5,10,20])
d = np.array([18,32,50])

cross_product = np.cross(c,d)

print(cross_product)
```

Projection of vector "a" on "v" vector:

```python
a = np.array([2,5])
v = np.array([8,-6])

#Magnitude of v
magnitude_of_v = np.sqrt(np.sum(v ** 2))

projection = (np.dot(a,v) / magnitude_of_v ** 2) * v

print(projection)
```
